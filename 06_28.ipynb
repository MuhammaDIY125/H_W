```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("hw.csv")
df.head()
```

### O'qish va yozish baxolari bir biri bilan qanday bog'liq?

```python
hue_colors = {'female': 'red',
              'male': 'blue'}

sns.scatterplot(x='math score',
                y='writing score',
                data=df,
                hue='gender',
                palette=hue_colors,
                alpha=0.4)
plt.show()
```

Betta ko'rishimiz mumkin ki matematikada yaxshi baxo olgan o'quvchilarni ko'pi yozishda ham yaxshi baxo oladi. Undan tashqari yana qiziq joylaridan biri bu qiz bollar asosan yozishdan yaxshi baxo olarkan, o'g'il bollar esa matematikadan.

```python
sns.relplot(x='math score',
            y='writing score',
            data=df,
            col='gender'
            )
plt.show()
```

### Qaysi guruhdagi o'quvchilar matematikadan yaxshi?

```python
sns.catplot(x='math score',
            y='race/ethnicity',
            data=df,
            kind='bar')

plt.show()
```

Ushbu barplot orqali biz **E** guruhidagi o'quvchilar matematika bo'yicha yaxshiroq o'qishini ko'rishimiz mumkin.

### Qaysi guruhdagi o'quvchilar o'qishdan yaxshi?

```python
sns.catplot(x='race/ethnicity',
            y='reading score',
            data=df,
            kind='box')
plt.show()
```

O'qish bo'yicha eng yaxshi natijalarni **C** va **E** guruhlar ko'rsatgan ekan, lekin oz bo'lsada **E** guruh yaxshiroq natijalar ko'rsatibti.

```python
sns.catplot(x='race/ethnicity',
            y='reading score',
            data=df,
            kind='box',
            whis=[0, 100])
plt.show()
```

Mana bu kodda yaqqolroq ko'rinyabti farqi

```python
sns.catplot(x='race/ethnicity',
            y='reading score',
            data=df,
            kind='bar')
plt.show()
```
