## ***Salary of Data Professions***

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('hw.csv')
```

### **Yosh bilan oylikni bog'liligi bormi?** 

```python
sns.regplot(data=df,
            x='AGE',
            y='SALARY')

plt.xlabel('AGE')
plt.ylabel('SALARY')
plt.grid(True)
plt.show()
```

Tekkis chiziq bizga xolatni yaxshi tasvirlolmayabti, egri chiziqda qilib ko'raylikchi

```python
sns.regplot(data=df,
            x='AGE',
            y='SALARY',
            order=2)
plt.show()
```

Ushbu **regplot** orqali ko'rishimiz mumkin ki, yosh kattalashgan sari oylik ham oshadi.

### **Nima uchun yosh oylikka bunday ta'sir qiladi?** 

```python
sns.scatterplot(data=df,
                x='AGE',
                y='PAST EXP',
                hue='SALARY')
plt.show()
```

```python
sns.regplot(data=df,
            x='AGE',
            y='PAST EXP',
            order=2)
plt.show()
```

Statistika bo'yicha yoshi katta bo'lgani sari eski ishida olgan opiti kattalasharkan, shuning uchun ham yoshi kattalarni oyligi kattaroq bo'lyabti, asosan 30 yoshdan kattalar hammasi malakalik ishchilar. Bunda 30 yoshdan kattalarni olish mumkin degan xulosaga kelishimiz mumkin.

### **Yosh va malakadan tashqari yana nimadir oylikka qattiq ta'sir qiladimi?** 

```python
c = ['AGE', 'SALARY', 'LEAVES USED', 'LEAVES REMAINING', 'RATINGS', 'PAST EXP']
df[c].corr()
```

```python
sns.heatmap(df[c].corr(), cmap='YlGnBu')
```
