## **Tozalash**

### Tozalashdan oldin ma'lumotlarni ko'rib chiqamiz

```python
df.info()
```

Ko'p column larda NaN kam ekan, lekin bazilarda (masalan SAT_AVG_ALL) NaN lar soni juda ko'p. Ularni yaxshirok ko'rib chiqishimiz kerak.

```python
c2=[]
for i in df.columns:
    if df[i].isna().sum():
        c2.append(i)
c=c2
print(df[c].isnull().sum())
len(c)
```

Umumiy bizda 15ta **NaN** larga ega column bor ekan. Ularni ichida jida ham ko'p **NaN** lik (masalan ADM_RATE_ALL), va kam **NaN** lik (masalan DEBT_MDN) lar ham bor.

```python
df[c]
```

Bu columnlarni alohida olib o'rganib chiqamiz

```python
df[c].info()
```

5-ta object va 10-ta float64 bor ekan. Qizig'i object tiplilardan 3tasi uzida sonlar saqlayabti, demak ularni konvert qilvorishimiz mumkin, omo bu columnlarda **PrivacySuppressed** degan joylari ham bor, ularni NaN qilib ketishimiz mumkin.

```python
box = ['COMPL_RPY_5YR_RT', 'DEBT_MDN', 'MEDIAN_HH_INC']

for i in box:
    df[i] = pd.to_numeric(df[i], errors='coerce')

df[box].info()
```

Muammoni xal qildik. Endi NaN ko'p qatnashgan olib tashlash mumkin bo'lgan ustunlarni olib tashlaymiz, qolganlariga esa mos qiymatlarni qo'yib chiqamiz.

```python
c2=[]
for i in c:
    if df.shape[0] / 10 > df[i].isnull().sum():
        c2.append(i)
    else:
        df = df.drop(i, axis=1)
c=c2
print(df[c].isnull().sum())
len(df.columns)
```

*24 columndan 16 qoldi. Bu katta yo'qatish, lekin juda ko'p NaN larga ega bo'lgan ustunlar bilan yaxshi taxlil qilib bo'lmaydi.

```python
df[c].info()
```

Ko'rishimiz mumkin ki 2ta object tipli columnlar qolibti. Oldin shularni NaNlarini tozalaymiz.

```python
for i in ['Locales', 'Locale_Short']:
    maska = df.groupby('REGION')[i].transform(pd.Series.mode)
    df[i] = df[i].fillna(maska)
    c.remove(i)
```

tipi categoriya bo'lgani uchun uni modasi bo'yicha almashtirdik. Har hil Region larda har hil bo'lganligi uchun esa Region buyicha gruhladik

```python
for i in c:
    maska = df.groupby(['REGION', 'Ownership'])[i].transform('median')
    df[i] = df[i].fillna(maska)
```

Qolgan numerik bo'lgan columnlarga esa REGION va Ownership bo'yicha guruhlab medianaga almashtirdik
